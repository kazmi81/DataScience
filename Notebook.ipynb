```python
Persons = {}

for i in range(2):
    name = input("Enter name for person "+str(i+1))
    age = int(input("Enter age for person "+str(i+1)))

    Persons[name] = age

older=max(Persons, key=Persons.get)
younger=min(Persons, key=Persons.get)

print(older +" is older than "+younger)
print(younger +" is younger than "+older)
```
